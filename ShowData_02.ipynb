# Weather Map Visualization (dynamic / realtime updates, A6)

```python
#start with libraries
import time
from visualization import StaticWeatherMapVisualizer, DynamicWeatherMapVisualizer
from Kafka import KafkaReader
import uuid
```

```python
# initialize kafka consumer for topic "weather.forecast.clean"
kafka_consumer = KafkaReader(bootstrap_servers='kafka-1,kafka-2', topic='weather.forecast.clean', group_id=uuid.uuid4().hex, client_id=uuid.uuid4().hex, auto_offset_reset='earliest')
vis = DynamicWeatherMapVisualizer(kafka_con=kafka_consumer)
vis.create_temperature_map() # create dynamic temperature map
vis.create_icon_map() # create dynamic icon map
vis.create_wind_map() # create dynamic wind map
while True:
    vis.update_map_realtime() # update all maps created with vis
    time.sleep(15*60) # e.g. every 15 Minutes (same interval as new data comes in)
```
